#### 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Максимизация точности (Precision):

1. Биометрические данные.
 
   Здесь важна точность, так как нельзя допустить ошибку при идентификации человека.
   
   
2. Сортировка особей.

   К примеру надо отсортировать хищных рыб от растительноядных.
   
   Надо выбрать максимально подходящих, чтобы наверняка сохранить популяцию (хоть и в пределах закрытого водоёма)                  растительноядных.

Максимизация полноты (Recall):
    
 1. Тестирование на болезнь/вирус и т.п. с высоким уровнем заражаемости  (лучше получить ложноположительный результат и изолировать здорового человека, чем получить ложноотрицательный и отпустить потенциально опасного человека).
 
 
 2. Отслеживание подозрительных транзакций в банке (лучше приостановить сомнительную транзакцию до выяснения обстоятельств, чем допустить утечку денежных средств мошенникам).

#### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-мера - метрика объединяющая в себе точность и полноту, выдавая как результат их гармоническое среднее, а коэффициент $\beta$, позволяет балансирвоать вклад, вносимый точностью, тем самым акцентируя внимание на той метрике, максимизация которой для нас наиболее приоритетна.

F-мера дает одинаковый вес и точности, и полноте, поэтому F-мера будет падать одинаково при уменьшении как точности, так и полноты. При оценке F-мерой важно соблюсти баланс между точностью и полнотой, поэтому, если одна из характеристик (Precision или Recall) будет стремиться к 0, а другая к 1, то общая F-мера будет стремиться к 0.
А среднее от Precision и Recall, будет просто стремиться к 0,5.

Если быть точнее, то в примере с F-мерой мы используем среднее, но не арифметическое, а гармоническое.
Среднее арифметическое A (среднее в обычном смысле) и среднее гармоническое H определяются следующим образом:

$$A =\frac{1}{n}\sum_{k=1}^n x_i = \frac{1}{n}(x_1 + x_2 + x_3 + \dots + x_n)$$$$H =\frac{n}{\sum_{k=1}^n \frac{1}{x_i}} = \frac{n}{(\frac{1}{x_1} + \frac{1}{x_2} + \frac{1}{x_3} + \dots + \frac{1}{x_n})}$$

Пусть $x1 = Precision$ и $x2 = Recall$, тогда $A$ и $H$ будут:

$$A = \frac{1}{2}(𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 + 𝑅𝑒𝑐𝑎𝑙𝑙)$$$$H = \frac{2}{(\frac{1}{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛} + \frac{1}{𝑅𝑒𝑐𝑎𝑙𝑙})} = \frac{2}{\frac{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 + 𝑅𝑒𝑐𝑎𝑙𝑙}{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 \cdot 𝑅𝑒𝑐𝑎𝑙𝑙}} = \frac{2 \cdot 𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 \cdot 𝑅𝑒𝑐𝑎𝑙𝑙}{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 + 𝑅𝑒𝑐𝑎𝑙𝑙} $$

Среднее гармоническое является более интуитивным, чем среднее арифметическое. Предположим, что у нас есть система распознавания отпечатков пальцев и ее точность и полнота будет 1,0 и 0,2 соответственно. Интуитивно понятно, что общая производительность система очень низкая, так как покрывает только 20% зарегистрированных отпечатков пальцев. При этом среднее арифметическое точности и полноты составляет 0,6, а среднее гармоническое:

$$\frac{2 \cdot 1 \cdot \frac{2}{10}}{1 + \frac{2}{10}} = \frac{4}{12} = \frac{1}{3}$$

Видно, что среднее гармоническое (0,333) является более разумной оценкой, чем среднее арифметическое (0,6).
